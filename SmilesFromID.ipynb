```python
import pandas as pd

```

```python
data = pd.read_csv("ids.csv")
```

```python
data.head()
```

```python
from urllib.request import urlopen
from urllib.parse import quote

def CIRconvert(ids):
    try:
        url = 'http://cactus.nci.nih.gov/chemical/structure/' + quote(ids) + '/smiles'
        ans = urlopen(url).read().decode('utf8')
        return ans
    except:
        return 'Did not work'

identifiers  = data["Name"]

```

```python
for ids in identifiers :
    smiles = (ids, CIRconvert(ids))
    data = data.append({'Smiles' : smiles}, ignore_index = True)
    data["Smiles"] = pd.Series(smiles)
```

```python
data.to_csv('Smiles.csv')
```
